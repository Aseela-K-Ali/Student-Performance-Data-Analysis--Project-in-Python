<h1 style="text-align:center;color:black;background-color:#B2DFDB;padding:10px;border-radius:8px;font-size:28px;">Data Preprocessing</h1>

This step ensures the dataset is clean,consistent,and suitable for analysis.The clening steps include **(i) Standardizing column names (ii) Creating derived columns (iii) Dropping unwanted columns**

### (i) Standardizing column names

```python
# Column name standardization
df.columns
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
df.columns
```

**Info:-** Column names are listed and removing leading spaces and converting them to lowercase.**str.strip()** method is used remove leading and tailing spaces from column names.Underscores are replaced with spaces using **str.replace(" ", "_")** and **str.lower()** converted each column name into lowercase for consistency.

### (ii) Creating derived columns

 In this step, new columns such as **total_grade_average,** **Performance_label,** **parents_edu_total,** **total_alc** are created to organize the data into clearer categories for analysis.

#### a) Calculating Overall Performance
To get a better understanding of a student's overall academic success, we created a new feature called **total_grade_average.** 
1. Calculating the mean of grades from all three stages (G1, G2, and G3).
2. Rounding the result to the nearest whole number.
3. Converting the data type to integer for consistency.

```python
df['total_grade_average']=(df['g1']+df['g2']+df['g3'])/3
df['total_grade_average']=df['total_grade_average'].round(0)
df['total_grade_average']=df['total_grade_average'].astype(int)
df
```

#### b) Categorizing Student Performance 

In this step,a new feature called **'Performance_label'**  is created by categorizing students based on their academic performance. Students are grouped into 'Above Average' (mark >= 15), 'Average' (mark >= 10), and 'Below Average' for better analysis and visualization.

```python
def performance_grade(mark):
    if mark>=15:
        return'Above Average'  
    elif  mark>=10:
        return 'Average'
    else:
        return 'Below Average'
df['Performance_label']=df['total_grade_average'].apply(performance_grade)
df
    
```

#### c) Combining Parental Education

In this step, we created a new feature called **'parents_edu_total'** by calculating the sum of the mother's education level (medu) and the father's education level (fedu).This combined score provides a better understanding of the overall academic environment at the student's home.

```python
df['parents_edu_total']=df['medu']+df['fedu']
df
```

#### d) Combining Alcohol Consumption
 
 In this step, we created a new feature called **'total_alc'** by combining daily alcohol consumption (dalc) and weekend alcohol consumption (walc).This combined metric helps to evaluate the overall impact of alcohol usage on a student's academic performance.

```python
df['total_alc']=df['dalc']+df['walc']
df
```

### (i) Dropping Unwanted Column

```python
df.drop(columns=['reason','nursery','romantic','pstatus','famsize','guardian'],inplace=True)
df
```

In this step, Removed 6 columns because they had very low crrelation with the students final performance .This process helped in reducing the complexity of the dataset.

### 1) Univariate Analysis

It examines each feature individually to understand it's distribution  and patterns

**i) Distribution of Student Grades** 

In this univariate analysis, I am examining the **'total_grade_average'** column to understand the overall academic performance of the students. I have used a **Histogram** to visualize the frequency distribution of these grades. This helps to identify the most common scores and see how the grades are spread across the entire dataset.

```python
#Histogram – Frequency distribution of Grades
plt.figure(figsize=(12,6))
plt.hist(df['total_grade_average'],bins=20,color='lightgreen',edgecolor='black')
plt.title(" Overall Students Performance ")
plt.xlabel("Average Grade")
plt.ylabel("Number Of Students")
plt.show()
```

**Insights:-** This Histogram represents the frequency distribution of student grades.It shows that the majority of students achieve average scores,with the highest concentration between 10 and 15.While the distribution follows a bell-like shape, there are very few students scoring at the extreme high  or low ends.

**ii) Distribution of Students by Rsidence(Urban vs. Rural)**

In this analysis, I have used the **'address'** column to understand the residential background of the students. I chose a **Pie Chart** for this univariate analysis to visualize the proportion of students living in Urban (U) and Rural (R) areas. This helps in identifying whether the dataset is more representative of a city-based or village-based student population.

```python
plt.figure(figsize=(10,6))
plt.title("Number of Students: Urban VS Rural")
address_counts=df['address'].value_counts()
plt.pie(address_counts,labels=['Urban(U)','Rural(R)'],autopct=lambda p: '{:.0f}'.format(p*address_counts.sum()/100),explode=(0,0.1),startangle=140,shadow=True,colors=['yellow','blue'])
plt.show()
```

**Insights:-** This pie chart illustrates the residential distribution of students. It is evident that the majority of students (452) come from Urban (U) areas, while a significantly smaller portion (197) reside in Rural (R) areas. This suggests that urban students represent about 70% of the total student population in this study.

**iii) Categorical Distribution of Students Performance**

I have used a **Count Plot** to perform this univariate analysis, visualizing how students are distributed across different performance levels. By using the **'Performance_label'** column, I have categorized the students into three groups: Below Average, Average, and Above Average. This chart helps to identify the overall performance trend within the student population at a glance.

```python
plt.figure(figsize=(10,8))
plt.xlabel('Performance Label')
plt.ylabel('Count of Students')
sb.countplot(x='Performance_label',data=df,palette='coolwarm',hue='Performance_label',legend=True)
plt.title(" Total Number Of Students in Each performance Category")
plt.show()
```

**Insights:-** The count plot shows that the majority of students fall into the 'Average' performance category.This distribution shows that only a small percentage of students from these schools achieve an 'Above Average ' level.

### 2) Bivariate Analysis

Bivariate analysis is the study of the relationship between two variables. We use it to see how one factor  impacts another.

**i) Impact of Study Time on Student Grades**

```python
# Study Time and Marks
plt.figure(figsize=(12,8))
sb.boxplot(x='studytime',y='total_grade_average',data=df)
plt.title("Impact of Study Time on Student Grades ")
plt.xlabel("Study Time")
plt.ylabel("Overall Average Grade")
plt.show()
```

**Insights:-** This box plot shows a positive correlation between study hours and accademic success.As the study hours increase, the median grade also shows a study rise.We can also observe some outliers in the lower study time groups, representing students who achieved high marks despite fewer study hours.

### 3) Multivariate Analysis

Multivariate Analysis is a statistical method used to analyze the relationship between three or more variables simultaneously. It helps to understand how multiple factors combined together impact a single outcome.

**i) Correlation Matrix of Student Performance**

"I have performed a Multivariate Analysis using a Correlation Heatmap from the Seaborn library. This chart shows how travel time, free time, and going out together impact the total grade average. The numbers in the boxes (Correlation) tell us how strongly these factors are related to the student's marks.

```python
data1=df[['total_grade_average','traveltime','freetime','goout']]
data1
```

```python
correlation=data1.corr()
correlation
```

```python
plt.figure(figsize=(10,6))
sb.heatmap(correlation,annot=True)
plt.title("Correlation Heatmap: Student Performance Factors")
plt.show()
```

**Insights:-** The analysis shows a negative correlation between grades and lifestyle factors like travel time, free time, and going out. This indicates that as these activities increase, the student's academic performance tends to decrease.    

**ii) Combined Influence of School and Family Support on Student Grades**

 Used a Grouped **Bar Plot** from the **Seaborn** library to perform this Multivariate Analysis. It compares how **school support** and **family support** together influence the student's average grades.

```python
plt.figure(figsize=(10,6))
plt.xlabel('School Support(Yes/No)')
plt.ylabel('Average Total Grade')
plt.title("Impact of School and Family Support on Grades")
support_analysis=df.groupby(['schoolsup','famsup'])['total_grade_average'].mean().unstack()
sb.barplot(x='schoolsup',y='total_grade_average',hue='famsup',data=df,errorbar=None)
plt.legend(title='Family Support',bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

**Insights:-** Students who receive both school and family support achieve the highest average grades. This visualization proves that a supportive environment at home (Family Support) and academic assistance at school are key factors in improving student performance.
